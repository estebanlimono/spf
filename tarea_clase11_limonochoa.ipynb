1. Usando el método de la transformada inversa.

```python
N = 500  # número de puntos

def distr_triang(x,a,b,c):
    if x >= a and x<=b:
        fx =  (x-a)*(2/((c-a)*(b-a)))
    elif x> b and x<=c:
        fx = (x-c)*(-2/((c-a)*(c-b)))
    else:
        fx = 0
    return fx
```

```python
# Función objetivo
x = np.arange(0,6,0.1)
a,b,c=1,2,5
f = list(map(lambda x: distr_triang(x,a,b,c) , x ))
# for i in range(0,N):
#     u1 = np.random.uniform(0,.5,1)  # valores de y 
#     u2 = np.random.uniform(1,5,1)  # valores de x
#     if u1 <= distr_triang(u2,a=1,b=2,c=5):
#         plt.plot(u2,u1,'ob')
#         i+= 1
#     else:
#         plt.plot(u2,u1,'or')
plt.title("Función creada")
plt.plot(x,f)
plt.show()
```

```python
# Función objetivo
x = np.arange(0,6,0.1)
a,b,c=1,2,5
f = list(map(lambda x: distr_triang(x,a,b,c) , x ))
for i in range(0,N):
    u1 = np.random.uniform(0,.5,1)  # valores de y 
    u2 = np.random.uniform(1,5,1)  # valores de x
    if u1 <= distr_triang(u2,a=1,b=2,c=5):
        plt.plot(u2,u1,'ob')
        i+= 1
    else:
        plt.plot(u2,u1,'or')
plt.title("Función creada")
plt.plot(x,f)
plt.show()
```
