```python
# Importamos las librerías principales
import numpy as np
import math as mt
import matplotlib.pyplot as plt
from scipy import optimize
from scipy.stats import beta
```

###  a). Caso particular: $\alpha=\beta=3$
Con estos valores la PDF es 
$$f(x)=30(x^2-2x^3+x^4)$$
sin pix

```python
N = 500  # número de puntos
# Función objetivo
f = lambda x: 30*(x**2-2*x**3+x**4)

x = np.arange(0,1,0.01)
c = float(f(optimize.fmin(lambda x:-f(x),0,disp=False)))
t=lambda x: c*np.ones(len(x))
plt.plot(x,f(x),label='f(x)')
plt.plot(x,t(x),label='t(x)')
print('El máximo de la función es:',c)
for i in range(N):
    u1 = np.random.rand(1)
    u2 = np.random.uniform(0,c,1)
    if u2 <= f(u1):
        plt.plot(u1,u2,'ob')
    else:
        plt.plot(u1,u2,'o')
plt.legend()        
plt.show()
```

```python
N = 500
f = lambda x: 30*(x**2-2*x**3+x**4)
c = f(optimize.fmin(lambda x: -f(x),0,disp=False))
t = lambda x: c*np.ones(len(x))
x = np.arange(0,1,0.01)
plt.plot(x,f(x),label='f(x)')
plt.plot(x,t(x),label='t(x)')
plt.legend()
R1 = np.random.rand(N)
R2 = np.random.rand(N)
f_x = f(R1)
t_x = t(R2)
condicion = np.multiply(R2,t_x) <= f_x
for i in range(N):
    if condicion[i]:
        plt.plot(R1[i],R2[i]*t_x[i],'ob')
    else:
        plt.plot(R1[i],R2[i]*t_x[i],'o')
plt.show()
```

```python
N = 500
f = lambda x: 30*(x**2-2*x**3+x**4)
c = f(optimize.fmin(lambda x: -f(x),0,disp=False))
t = lambda x: c*np.sin(np.pi*x)
x = np.arange(0,1,0.01)
plt.plot(x,f(x),label='f(x)')
plt.plot(x,t(x),label='t(x)')
plt.legend()
R1 = np.random.rand(N)
R2 = np.random.rand(N)
f_x = f(R1)
t_x = t(R1)
condicion = np.multiply(R2,t_x) <= f_x
for i in range(N):
    if condicion[i]:
        plt.plot(R1[i],R2[i]*t_x[i],'ob')
    else:
        plt.plot(R1[i],R2[i]*t_x[i],'or')
plt.show()
```
